## Tarefa 01:

* Sabendo que no processo de aprendizagem de um modelo é necessário utilizar parte da base de dados, faça testes em uma cópia do modelo **Atividade 01** alterando as porcentagens (porções) utilizadas na fase de treinamento:

  * Teste 01: 25%
  * Teste 02: 50%
  * Teste 03: 75%
  * Teste 04: 90%

  Qual teste apresentou maior taxa de acerto (R2)?

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

url = 'https://raw.githubusercontent.com/CarlosR-SantosJr/DisciplinaInteligenciaArtificial-INTI6/main/datasets/HousingData.csv'
df = pd.read_csv(url)

colunas = df.columns
imputer = SimpleImputer()
df = pd.DataFrame(imputer.fit_transform(df), columns=colunas)

X = pd.DataFrame(df['MEDV'])
y = pd.DataFrame(df['LSTAT'])

resultados = []

for percentual in [0.25, 0.50, 0.75, 0.90]:
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, train_size=percentual, random_state=4
    )

    modelo = LinearRegression()
    modelo.fit(X_train, y_train)

    y_pred = modelo.predict(X_test)
    r2 = r2_score(y_test, y_pred)

    resultados.append({
        'percentual_treino': percentual,
        'r2': r2
    })

resultados_df = pd.DataFrame(resultados)
resultados_df['percentual_treino'] = (
    (resultados_df['percentual_treino'] * 100).astype(int).astype(str) + '%'
)
resultados_df = resultados_df.sort_values(by='r2', ascending=False).reset_index(drop=True)

print(resultados_df)
print(
    'Melhor resultado:',
    resultados_df.loc[0, 'percentual_treino'],
    'com R2 =',
    round(resultados_df.loc[0, 'r2'], 4)
)
```
